```python
x1 = df[df['Species']=='Iris-setosa']['Species']
y1 = df[df['Species']=='Iris-setosa']['PetalLengthCm']
x2 = df[df['Species']=='Iris-virginica']['Species']
y2 = df[df['Species']=='Iris-virginica']['PetalLengthCm']
x3 = df[df['Species']=='Iris-versicolor']['Species']
y3 = df[df['Species']=='Iris-versicolor']["PetalLengthCm"]
```

```python
plt.scatter(y1, x1, marker='+',label="Setosa")
plt.scatter(y3, x3, marker='o',label="Versicolor")
plt.scatter(y2, x2, marker='*',label="Virginica")
plt.xlabel("PetalLenthCm")
plt.ylabel("Species")
plt.legend()
plt.show()
```
